# 调整图形：坐标上下限

### 虽然 Matplotlib 会自动为你的图形选择最合适的坐标轴上下限，但是有时自定义坐标轴上下限可能会更好。调整坐标轴上下限最基础的方法是 plt.xlim() 和 plt.ylim()

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
plt.show()
```

```python
#如果你想要让坐标轴逆序显示，那么也可以逆序设置坐标轴刻度值：
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x))
plt.xlim(10, 0)
plt.ylim(1.2, -1.2);
plt.show()
```

#### 有时在二维图上用等高线图或者彩色图来表示三维数据是个不错的方法。Matplotlib 提供了三个函数来解决这个问题：用 plt.contour 画等高线图、用 plt.contourf 画带有填充色的等高线图（filled contour plot）的色彩、用 plt.imshow 显示图形

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

## 用等高线图可视化三维数据

```python
def f(x,y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
#等高线图可以用 plt.contour 函数来创建。它需要三个参数：x 轴、y 轴、z 轴三个坐标轴的网格数据
#x 轴与 y 轴表示图形中的位置，而 z 轴将通过等高线的等级来表示。
#用np.meshgrid 函数来准备这些数据可能是最简单的方法，它可以从一维数组构建二维网格数据：
x = np.linspace(0, 5, 50)#创建一维数组
y = np.linspace(0, 5, 40)
X, Y = np.meshgrid(x, y)#将两个一维数组网格化
Z = f(X, Y)#将网格化的数据与Z相关联
```

```python
plt.contour(X, Y, Z, colors='black')#用等高线图可视化三维数据
```

```python
#需要注意的是，当图形中只使用一种颜色时，默认使用虚线表示负数，使用实线表示正数。
#你可以用 cmap 参数设置一个线条配色方案来自定义颜色。
#还可以让更多的线条显示不同的颜色——可以将数据范围等分为 20 份，然后用不同的颜色表示：
plt.contour(X, Y, Z, 20, cmap='RdGy')#RdGy （红 - 灰，Red-Gray 的缩写）
#Matplotlib 有非常丰富的配色方案，可以在 IPython 里用 Tab 键浏览 plt.cm 模块对应的信息：
```

# 颜色条刻度的限制与扩展功能的设置

##### Matplotlib 提供了丰富的颜色条配置功能。由于可以将颜色条本身仅看作是一个 plt.Axes实例，因此所有关于坐标轴和刻度值的格式配置技巧都可以派上用场。颜色条有一些有趣的特性。例如，我们可以缩短颜色取值的上下限，对于超出上下限的数据，通过 extend 参数用三角箭头表示比上限大的数或者比下限小的数。这种方法很简单，比如你想展示一张噪点图

```python
import matplotlib.pyplot as plt
plt.style.use('classic')
import numpy as np
x = np.linspace(0, 10, 1000)
I = np.sin(x) * np.cos(x[:, np.newaxis])
# 为图形像素设置1%噪点
speckles = (np.random.random(I.shape) < 0.01)#I的shape是(1000,1000),
I[speckles] = np.random.normal(0, 3, np.count_nonzero(speckles))#生成一个正态分布的数组
plt.figure(figsize=(10, 3.5))
plt.subplot(1, 2, 1)#三个参数分别是行数、列数、索引值
plt.imshow(I, cmap='RdBu')
plt.colorbar()
plt.subplot(1, 2, 2)
plt.imshow(I, cmap='RdGy')
plt.colorbar(extend='both')#扩展图形
plt.clim(-1,1)#设置了刻度的上下限,只能用在imshow后，控制渲染后的图形
plt.show()
```

# 离散型颜色条

### 虽然颜色条默认都是连续的，但有时你可能也需要表示离散数据。最简单的做法就是使用plt.cm.get_cmap() 函数，将适当的配色方案的名称以及需要的区间数量传进去即可

```python
import matplotlib.pyplot as plt
plt.style.use('classic')
import numpy as np
x = np.linspace(0, 10, 1000)
I = np.sin(x) * np.cos(x[:, np.newaxis])
# 为图形像素设置1%噪点
speckles = (np.random.random(I.shape) < 0.01)
I[speckles] = np.random.normal(0, 3, np.count_nonzero(speckles))
plt.imshow(I, cmap=plt.cm.get_cmap('Blues', 8))#这里是设置颜色分布的区间和需要的区间数量
plt.colorbar()
plt.clim(-1, 1);
plt.show()
```

```python
#有三种方式解决这种问题
#ax.transData	以数据为基准的坐标变换。
#ax.transAxes	以坐标轴为基准的坐标变换（以坐标轴维度为单位）。
#fig.transFigure	以图形为基准的坐标变换（以图形维度为单位）。
```

#### transAxes 坐标以坐标轴（图中白色矩形）左下角的位置为原点，按坐标轴尺寸的比例呈现坐标。
#### transFigure 坐标与之类似，不过是以图形（图中灰色矩形）左下角的位置为原点，按图形尺寸的比例呈现坐标。

## 3.3极坐标曲面图

### 画曲面图需要二维网格化数据，但可以不是直角坐标系

#### 下面用一个极坐标网格来画曲面图

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
def f(x,y):
    return np.sin(np.sqrt(x**2+y**2))
r=np.linspace(0,6,20)
theta=np.linspace(-0.9*np.pi,0.8*np.pi,40)
r,theta=np.meshgrid(r,theta)
X=r*np.sin(theta)
Y=r*np.cos(theta)
Z=f(X,Y)
fig=plt.figure(figsize=(20,8),dpi=80)
ax=plt.axes(projection='3d')
ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap='viridis',edgecolor='none')
plt.show()
```

# 案例 莫比乌斯带

```python
theta=np.linspace(0,2*np.pi,30)
w=np.linspace(-0.25,0.25,8)
w,theta=np.meshgrid(w,theta)
phi=0.5*theta
#x-y平面内的半径
r=1+w*np.cos(phi)
x=np.ravel(r*np.cos(theta))
y=np.ravel(r*np.sin(theta))
z=np.ravel(w*np.sin(phi))
#用基本参数化方法定义三角剖分
from matplotlib.tri import Triangulation
tri=Triangulation(np.ravel(w),np.ravel(theta))
fig=plt.figure(figsize=(10,8),dpi=80)
ax=plt.axes(projection='3d')
ax.plot_trisurf(x,y,z,triangles=tri.triangles,cmap='viridis',linewidths=0.2)
ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
ax.set_zlim(-1,1)
plt.show()
plt.savefig('莫比乌斯带.png')
```
